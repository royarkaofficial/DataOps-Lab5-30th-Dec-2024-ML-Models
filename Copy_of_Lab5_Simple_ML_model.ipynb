```python
import pandas as pd
```

Load the dataset

```python
data = pd.read_csv("/content/sample_data/synthetic_iris_data.csv")
```

```python
data.head()
```

```python
x = data.iloc[:,:-1]
y = data.iloc[:,-1]
```

```python
x
```

```python
y
```

Split the data into Training and Testing

```python
from sklearn.model_selection import train_test_split

X_train , X_test , y_train , y_test = train_test_split (x, y , test_size = 0.2 , random_state = 42)
# x = features , y = target
```

```python
# Select the regression algo

from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
```

```python
#Initiate the model training
model.fit(X_train , y_train)
```

```python
X_test , y_test
```

```python
#Test the trained model

ypred = model.predict(X_test)
```

```python
#Calculate the accuracy with the help of Actual value(y_test) and predicted values (y_pred)
from sklearn.metrics import accuracy_score
acc = accuracy_score(y_test , ypred)
print(acc)
```

```python
new_data = [5.1,6.5,3.2,1.8]
```

```python
op = model.predict([new_data])
```
